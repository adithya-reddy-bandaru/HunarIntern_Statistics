```python
import matplotlib.pyplot as plt

years = [1810, 1820, 1830, 1840, 1850, 1860, 1870, 1880, 1900, 1910, 1920, 
         1930, 1940, 1950, 1960, 1970, 1980]
heights = [169.7, 169.1, 166.7, 166.5, 165.6, 166.6, 167.2, 168.0, 169.4, 
           170.9, 171.0, 173.9, 174.9, 176.0, 176.9, 177.1, 176.8]

mean_height = sum(heights) / len(heights)

print(f"Mean height: {mean_height:.2f} cm")

plt.figure(figsize=(10, 6))
plt.plot(years, heights, marker='o', linestyle='-', color='b', label='Height (cm)')
plt.title('Heights of Males in the UK (1800s - 1980)')
plt.xlabel('Year')
plt.ylabel('Height (cm)')
plt.xticks(years, rotation=45)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
